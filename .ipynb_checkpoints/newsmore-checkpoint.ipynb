```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
## Importing dataset and cleaning it

df = pd.read_csv("headlines_data.csv")
df = df[:50000]
df = df["headline_text"]
headlines = df.tolist()
```

```python
headlines[:3]
```

```python
## Creating a bigram model for the headlines

b = {}

for headline in headlines:
    chs = ["<S>"] + list(headline) + ["<E>"]
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        b[bigram] = b.get(bigram, 0) + 1
```

```python
## Sorting the bigrams according to the occurance

# sorted(b.items(), key= lambda kv: -kv[1])
```

```python
chars = sorted(list(set("".join(headlines))))
s_to_i = {s: i for i, s in enumerate(chars)}
s_to_i["<B>"] = 38
```

```python
N = torch.ones((39, 39), device=device, dtype=torch.int32)     ## Had started with zeros first, then started with ones to avoid infinite log-loss problem
```

```python
for headline in headlines:
    chs = ["<B>"] + list(headline) + ["<B>"]
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = s_to_i[ch1]
        ix2 = s_to_i[ch2]
        N[ix1, ix2] += 1
```

```python
i_to_s = {i: s for s, i in s_to_i.items()}
```

```python
plt.figure(figsize=(20, 20))
plt.imshow(N.cpu().detach().numpy(), cmap="Blues")

for i in range(39):
    for j in range(39):
        ch_str = i_to_s[i] + i_to_s[j]
        plt.text(j, i, ch_str, ha="center", va="bottom", color="gray")
        plt.text(j, i, N[i, j].item(), ha="center", va="top", color="gray")

plt.axis("off")
```

```python
## Sampling one character using multinomial distribution

g = torch.Generator(device=device).manual_seed(1111)

p = N[0].float()
p = p/p.sum()
ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g)
i_to_s[ix.item()]
```

```python
P = N.float()
P /= P.sum(1, keepdim=True)
```

```python
P[0]
```

```python
## Generating a complete sentence using multinomial distribution

g = torch.Generator(device=device).manual_seed(1111)

for i in range(10):
    out = []
    ix = 0
    while True:
        # p = N[ix].float()
        # p = p/p.sum()
        p = P[ix]
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        
        if ix == 38:
            break;

        out.append(i_to_s[ix])

    print("".join(out))
```

```python
## Making the log-loss function

log_likelihood = 0.0
n = 0

for headline in headlines:
    chs = ["<B>"] + list(headline) + ["<B>"]
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = s_to_i[ch1]
        ix2 = s_to_i[ch2]
        prob = P[ix1, ix2]
        log_prob = torch.log(prob)
        log_likelihood += log_prob
        n += 1

print(f"{log_likelihood=}")
nll = -log_likelihood
print(f"{nll=}")
print(f"{nll/n}")
```

```python
# Following the Neural Network Approach

## Although it is named neural network approach, it is still gonna work on the bigram model where the model tries to predict the next character.s
```

```python
## Creating the training set: Here x is the current character and y is the next character which the model has to predict

x, y = [], []

for headline in headlines:
    chs = ["<B>"] + list(headline) + ["<B>"]
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = s_to_i[ch1]
        ix2 = s_to_i[ch2]
        x.append(ix1)
        y.append(ix2)

x = torch.tensor(x, device=device)
y = torch.tensor(y, device=device)

x, y
```

```python
x.shape
```

```python
## One-Hot Encoding these inputs cause the inputs needs to be normalized

x_enc = F.one_hot(x).float()
y_enc = F.one_hot(y).float()

x_enc.shape
```

```python
## Initialize the weights for the network

w = torch.randn(39, 39, device=device, generator=g, requires_grad=True)
```

```python
##

logits = x_enc @ w
counts = logits.exp()
probs = counts/counts.sum(1, keepdims=True)
p.shape
```

```python
## Training the model

epochs = 50
lr = 100
for epoch in range(epochs):
    
    ## Forward pass
    logits = x_enc @ w
    counts = logits.exp()
    probs = counts / counts.sum(1, keepdims=True)
    loss = -probs[torch.arange(2019342), y].log().mean()

    ## Backward pass
    w.grad = None
    loss.backward()

    ## Gradient Descent
    w.data += -lr * w.grad

    if epoch%10 == 0:
        print(f"epoch: {epoch} | loss: {loss.item():5f}")

print(f"epoch: {epoch+1} | loss: {loss.item():5f}")
```

```python
## Predicting the characters based on this Neural Network

for i in range(10):
    ix = 0
    out = []

    while True:
        x_enc_pred = F.one_hot(torch.tensor([ix]), num_classes=39).float()
        x_enc_pred = x_enc_pred.to(device)
        logits = x_enc_pred @ w
        counts = logits.exp()
        p = counts / counts.sum(1, keepdims=True)

        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out.append(i_to_s[ix])

        if ix == 38:
            break

    print("".join(out))
```
